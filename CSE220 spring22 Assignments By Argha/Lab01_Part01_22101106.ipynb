```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
    for i in range(len(source)-k):
        source[i] = source[i+k]
    for i in range(k):
        source[-i-1] = 0
  # Hint, You can write a function for left shift once and then use it
    return source



# Test 02: Rotate Left k cell
def rotate_left(source, k):
  # TO DO
    new = [0]*len(source)

    for i in range(len(source)):
        new[i] = source[(i+k) % len(source)]
  # Hint, You can write a function for left rotate once and then use it
    return new


# Test 03: Shift Right k cell
def shift_right(source, k):
  # TO DO
    for i in range(len(source)-k):
        source[i+k] = source[i]
    for i in range(k):
        source[i] = 0
  # Hint, You can write a function for right shift once and then use it
    return source



# Test 04: Rotate Right k cell
def rotate_right(source, k):
  # TO DO
    new = [0] * len(source)

    for i in range(len(source)):
        new[(i+k) % len(source)] = source[i]

  # Hint, You can write a function for right rotate once and then use it
    return new



# Test 05: Remove an element from an array
def remove(source, idx):
  # TO DO
    # Problem checking:
    if 0 > idx or idx >= len(source)-1:
        print("invalid index")
        return

    for i in range(len(source)-1-idx):
        source[idx+i] = source[idx+i+1]
    source[len(source)-1] = 0

    return source



# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
  # Problem checking:
    if element not in source:
        return

    # Code
    count = 0

    for i in range(len(source)):                  # needs more efficiency
        i = i-count
        if source[i] == element:
            for j in range(len(source)-1-i):
                source[j + i] = source[j + i + 1]
            source[len(source)-1] = 0
            count += 1

    return source



# Test 07: Splitting an Array
def split_array(a):
  # TO DO
    ans = False
    for i in range(len(a)):
        sum_first_half = 0
        for j in range(i):
            sum_first_half += a[j]

        sum_second_half = 0
        for j in range(i, len(a)):
            sum_second_half += a[j]

        if sum_first_half == sum_second_half:
            ans = True
            break

    return ans


# Test 08: Max Bunch Count
def max_bunch(source):
  # TO DO
    highest, count = 1, 1

    for i in range(len(source)-1):
        if source[i] == source[i + 1]:
            count += 1
        else:
            highest = count if count > highest else highest  # hudai, moja lge 1 liners
            count = 1

    return count
```

```python
def calculate_mean (source):
    sum = 0
    for i in source:
        sum += i
    return sum/len(source)

def calculate_standard_deviation(source):
    squared_list = [(i-calculate_mean(source))**2 for i in source]
    sum = 0
    for i in squared_list:
        sum += i

    return round((sum/(len(source)-1))**0.5,2)



def append(old, new_element):
    new = [0]*(len(old)+1)
    for i in range(len(old)):
        new[i] = old[i]
    new[len(new)-1] = new_element
    return new

def checking_standard_deviation_away(source, k):
    x_above = calculate_mean(source) + k*calculate_standard_deviation(source)
    x_below = calculate_mean(source) - k*calculate_standard_deviation(source)

    new = []
    for i in source:
        if i >= x_above or i <= x_below:
            new = append(new,i)

    return new






given = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
print(f"The mean of the numbers is: {calculate_mean(given)}")
print(f"The standard deviation is: {calculate_standard_deviation(given)}")
print(f"New array: {checking_standard_deviation_away(given, 1.5)}")


```
