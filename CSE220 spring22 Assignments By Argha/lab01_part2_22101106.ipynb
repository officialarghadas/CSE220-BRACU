```python
import math
sample=[10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]

def meancounting(sample):
  mn=0
  for i in range(len(sample)):
    mn+=sample[i]
  mean= mn/len(sample)
  return mean

def SDF(sample):
  mean=meancounting(sample)
  n=(len(sample)-1)
  total=0
  for i in range(len(sample)):
    var=(sample[i]-mean) **2
    total+=var
    var2=(total/n)
    sdf_final=math.sqrt(var2)
  return sdf_final

def sdf_counting(sample):
  sdf=SDF(sample)
  mean=meancounting(sample)
  newarrray=[]
  for i in range(len(sample)):
    neu1= mean + 1.5 * sdf
    neu2= mean - 1.5* sdf
    if sample[i]>neu1 or sample[i]< neu2:
      newarrray.append(sample[i])
  return newarrray
print(f"The mean of the numbers is:",meancounting(sample))
print(f"The standard deviation is: ",SDF(sample))
print(f"New array: ",sdf_counting(sample))
```
